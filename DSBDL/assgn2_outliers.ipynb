```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/home/pict/31162/academic-performance.csv')
```

```python
#checking available values
data.notnull()
```

```python
#copying dataset for operation
tempdata = data

#FILLING NULL VALUES
tempdata['Che_marks'].fillna(tempdata['Che_marks'].mean(),inplace = True)
tempdata['Phy_marks'].fillna(tempdata['Phy_marks'].mean(),inplace = True)
tempdata
```

```python
#boxplot shows the distribution of the data
col = [ 'Phy_marks','Che_marks','EM1_marks']
tempdata.boxplot(col)
```

```python
#removing negative values

for idx in tempdata.index :
    if(tempdata.loc[idx,'Phy_marks']<0):
        tempdata.loc[idx,'Phy_marks']=0

tempdata
```

```python
#deleting null values using dropna()

#To Drop rows if all values in that row are missing
tempdata.dropna(how='all')
tempdata
```

```python
# To Drop columns with at least 1 null value.
nds = tempdata.dropna(axis = 1)
nds
```

```python
# To Drop row with at least 1 null value.
nds1 = tempdata.dropna(axis = 0, how ='any')
nds1
```
