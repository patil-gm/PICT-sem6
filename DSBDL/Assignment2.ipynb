```python
# IQR
```

```python
q1 = np.percentile(df["Phy_marks"],25,interpolation='midpoint')
q3 = np.percentile(df["Phy_marks"],75,interpolation='midpoint')
iqr = q3-q1

phyupper = np.where(df["Phy_marks"]>=(q3+1.5*iqr))
phylower = np.where(df["Phy_marks"]<=(q1-1.5*iqr))
print("Phy_marks",phylower[0],phyupper[0])
```

```python
q1 = np.percentile(df["Che_marks"],25,interpolation='midpoint')
q3 = np.percentile(df["Che_marks"],75,interpolation='midpoint')
iqr = q3-q1

cupper = np.where(df["Che_marks"]>=(q3+1.5*iqr))
clower = np.where(df["Che_marks"]<=(q1-1.5*iqr))
print("Che_marks",clower[0],cupper[0])
```

```python
q1 = np.percentile(df["PPS_marks"],25,interpolation='midpoint')
q3 = np.percentile(df["PPS_marks"],75,interpolation='midpoint')
iqr = q3-q1

ppsupper = np.where(df["PPS_marks"]>=(q3+1.5*iqr))
ppslower = np.where(df["PPS_marks"]<=(q1-1.5*iqr))
print("PPS_marks",ppslower[0],ppsupper[0])
```

```python
df.drop(ppslower[0],inplace=True)
```

```python
df.drop(phylower[0],inplace=True)
```

```python
# Z-Score
```

```python
mean = np.mean(df["Che_marks"])
std = np.std(df["Che_marks"])
threshold = -2
outliers = []
```

```python
for i in df["Che_marks"]:
    z = (i-mean)/std
    if z<threshold:
        outliers.append(i)

print(outliers)
```
