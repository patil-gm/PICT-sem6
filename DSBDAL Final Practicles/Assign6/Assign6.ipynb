Problem Statement :  
Write a Python program to display some basic statistical details like percentile, mean, standard deviation etc. of the species of ‘Iris-setosa’, ‘Iris-versicolor’ and ‘Iris- verginica’ of iris.csv dataset. 
Grouped by one of the qualitative (categorical) variable from the dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('../Databases/iris.csv')
df
```

```python
# Group the data by species and calculate the statistics
grouped_data = df.groupby('Species').describe()
grouped_data
```

```python
# Select the relevant statistics for the grouped species
selected_stats = grouped_data['SepalLengthCm'][['mean', 'std', 'min', '25%', '50%', '75%', 'max']]
selected_stats
```
